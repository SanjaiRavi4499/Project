```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess data
#data = np.loadtxt('data.txt')
#data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

# Initialize empty arrays to hold silhouette scores and cluster sizes
sil_scores = []
cluster_sizes = []

# Loop through different numbers of clusters and calculate silhouette scores
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(data_cleaned)
    labels = kmeans.labels_
    score = silhouette_score(data_cleaned, labels, metric='euclidean')
    sil_scores.append(score)
    cluster_sizes.append(i)

# Plot silhouette scores for each number of clusters
plt.plot(cluster_sizes, sil_scores)
plt.title('Silhouette Scores for KMeans Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.show()

# Identify optimal number of clusters with highest silhouette score
optimal_cluster_size = cluster_sizes[np.argmax(sil_scores)]
print('Optimal number of clusters: ', optimal_cluster_size)
```

## 3.Calinski-Harabasz index

```python
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
import matplotlib.pyplot as plt

# Calculate Cainski-Harabasz index for different numbers of clusters
scores = []
for k in range(2,11,1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_cleaned)
    score = calinski_harabasz_score(data_cleaned, kmeans.labels_)
    scores.append(score)

# Plot elbow graph to determine optimal number of clusters
plt.plot(range(2,11,1), scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Calinski-Harabasz score')
plt.show()
```
